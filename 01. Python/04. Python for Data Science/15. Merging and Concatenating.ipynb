## Concatenation

```python
# dataframes having the same columns
df1 = pd.DataFrame({'Name': ['Afif', 'Nadim', 'Sanzeda', 'Masudul'],
                    'Age': [34, 30, 22, 33],
                    'Gender': ['M', 'M', 'F', 'M']}
                  )

df2 = pd.DataFrame({'Name': ['Kamrul', 'Vikram', 'Priya'],
                    'Age': [31, 22, 19],
                    'Gender': ['M', 'M', 'F']}
                  )
```

```python
df1
```

```python
df2
```

```python
df = pd.concat([df1, df2], axis = 0)
df= df.reset_index()
df = df.drop(['index'], axis = 1)
df
```

```python
df = df1.append(df2)
df = df.reset_index()
df = df.drop(['index'], axis = 1)
df
```
